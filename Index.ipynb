```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
```

```python
#constants
rho = 28
sigma = 10.0
beta = 8.0 / 3.0
state0 = [1.0, 1.0, 1.0]
n=4000
upperBound=(n+1)/100
```

```python
def f(state, t):
    x, y, z = state  # Unpack the state vector
    return sigma * (y - x), x * (rho - z) - y, x * y - beta * z  # Derivatives

#time array created for integration
t = np.arange(0.0, upperBound, 0.01)

#integrate the function f, with initial state state0, over the time interval t
states = odeint(f, state0, t)

x=states[:,0]
y=states[:,1]
z=states[:,2]
```

```python
##coloring array
colorChanger=np.linspace(0,1,np.size(x))**2
```

```python
#setting up figure
fig = plt.figure(dpi=1200)
ax = plt.axes(projection="3d")

#actually plotting graphs
s = 10
for i in range(0,n-s,s):
    colorMine=colorChanger[i]
    ax.plot(x[i:i+s+1],y[i:i+s+1],z[i:i+s+1], color=(0,0.5,colorMine))
#     print(colorMine)

#saving and displaying
# plt.axis('off')
# plt.savefig('background.png', format="png", transparent=True)
plt.draw()
plt.show()
```
